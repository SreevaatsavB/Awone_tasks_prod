```python
from matplotlib import pyplot as plt
```

```python
bsz = [i+1 for i in range(10)]

gpu_mem_fwd_pass = [0.2, 0.4,0.6, 0.7, 0.9, 1.1, 1.3, 1.5, 1.7, 1.8]

plt.figure()
plt.scatter(y = gpu_mem_fwd_pass, x = bsz)
plt.title("GPT-2 GPU memory consumption (fwd pass)")
plt.xlabel("Batch size")
plt.ylabel("GPU Mem (GB)")
plt.show()

bsz = [i+1 for i in range(10)]

gpu_mem_fwd_pass = [0.2, 0.7, 1, 1.5, 1.9, 2.4, 2.8, 3.3, 3.7, 4]

plt.figure()
plt.plot(bsz , gpu_mem_fwd_pass, "o-")

plt.title("GPT-2 GPU memory consumption (fwd pass)")
plt.xlabel("Batch size")
plt.ylabel("GPU Mem (GB)")
plt.show()
```
